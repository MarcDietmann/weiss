```python
# Data Science Utils
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, f1_score
from sklearn.utils import resample

# ML Models
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
```

```python
df_plc_7d = pd.read_csv('plc_dS_7d.csv')
```

```python
df_plc_7d
```

```python
df_plc_7d_cols = df_plc_7d[["_time", "_value", "_field"]]
df_plc_7d_cols.head()
```

```python
pivoted_df = df_plc_7d_cols.pivot(index='_time', columns='_field', values='_value').reset_index()
```

```python
pivoted_df['_time'] = pd.to_datetime(pivoted_df['_time'])
```

```python
pivoted_df.dtypes
```

```python
# Create a new figure with two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 6), sharex=True)

# Plot 'MaxLastCycle' in the first subplot (ax1)
ax1.plot(pivoted_df['_time'], pivoted_df['MaxLastCycle'], marker='o', linestyle='-', color='b')
ax1.set_ylabel('MaxLastCycle')
ax1.set_title('MaxLastCycle and RmsLastCycle Over Time')

# Plot 'RmsLastCycle' in the second subplot (ax2)
ax2.plot(pivoted_df['_time'], pivoted_df['RmsLastCycle'], marker='o', linestyle='-', color='g')
ax2.set_xlabel('Timestamp')
ax2.set_ylabel('RmsLastCycle')


# Rotate x-axis labels for better readability (optional)
plt.xticks(rotation=45)

# Customize x-axis ticks to show a label every 12 hours
ax1.xaxis.set_major_locator(plt.MaxNLocator(10))  # Adjust the number as needed
ax2.xaxis.set_major_locator(plt.MaxNLocator(10))

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
